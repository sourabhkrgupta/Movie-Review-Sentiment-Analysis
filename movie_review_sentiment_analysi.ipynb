## In this  Project, we’ll build binary classification that puts movie reviews texts into one of two categories — negative or positive sentiment.

Importing the necessary libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import re # for regex
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.metrics import accuracy_score
import pickle
import warnings
warnings.filterwarnings('ignore')
```

# No null values, Label encode sentiment to 1(positive) and 0(negative)

```python
data.sentiment.value_counts()
```

```python
data.sentiment.replace('positive',1,inplace=True)
data.sentiment.replace('negative',0,inplace=True)
data.head(10)
```

```python
data.review[0]
```

# Steps to Clean the Reviews:
1. Remove HTML tags
2. Remove special characters
3. Convert everything to lowercase
4. Remove stopwords
5. Stemming

# 1. Remove HTML tags

```python
def clean(text):
    cleaned = re.compile(r'<.*?>')
    return re.sub(cleaned,'',text)

data.review = data.review.apply(clean)
data.review[0]
```

# 2. Remove special characters

```python
def is_special(text):
    rem = ''
    for i in text:
        if i.isalnum():
            rem = rem + i
        else:
            rem = rem + ' '
    return rem

data.review = data.review.apply(is_special)
data.review[0]
```

# Convert everything to lowercase

```python
def to_lower(text):
    return text.lower()

data.review = data.review.apply(to_lower)
data.review[0]
```

# Remove stopwords

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
def rem_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text)
    return [w for w in words if w not in stop_words]

data.review = data.review.apply(rem_stopwords)
data.review[0]
```

# Stem the words

```python
def stem_txt(text):
    ss = SnowballStemmer('english')
    return " ".join([ss.stem(w) for w in text])

data.review = data.review.apply(stem_txt)
data.review[0]
```

```python
data.head()
```

# CREATING THE MODEL

# 1. Creating Bag Of Words (BOW)

```python
X = np.array(data.iloc[:,0].values)
y = np.array(data.sentiment.values)
cv = CountVectorizer(max_features = 1000)
X = cv.fit_transform(data.review).toarray()
print("X.shape = ",X.shape)
print("y.shape = ",y.shape)
```

```python
print(X)
```

# 2. Train test split

```python
trainx,testx,trainy,testy = train_test_split(X,y,test_size=0.2,random_state=9)
print("Train shapes : X = {}, y = {}".format(trainx.shape,trainy.shape))
print("Test shapes : X = {}, y = {}".format(testx.shape,testy.shape))
```

# Defininh the models and Training them

```python
gnb,mnb,bnb = GaussianNB(),MultinomialNB(alpha=1.0,fit_prior=True),BernoulliNB(alpha=1.0,fit_prior=True)
gnb.fit(trainx,trainy)
mnb.fit(trainx,trainy)
bnb.fit(trainx,trainy)
```

# 4. Prediction and accuracy metrics to choose best model

```python
ypg = gnb.predict(testx)
ypm = mnb.predict(testx)
ypb = bnb.predict(testx)

print("Gaussian = ",accuracy_score(testy,ypg))
print("Multinomial = ",accuracy_score(testy,ypm))
print("Bernoulli = ",accuracy_score(testy,ypb))
```

## Define prediction function


Function that takes a text review as input, preprocesses it, transforms it using the loaded `CountVectorizer`, and returns the predicted sentiment (Positive or Negative) using the loaded model.


```python
def predict_sentiment(text):
    # Apply preprocessing steps
    cleaned_review = clean(text)
    special_char_removed_review = is_special(cleaned_review)
    lowercase_review = to_lower(special_char_removed_review)
    stopwords_removed_review = rem_stopwords(lowercase_review)
    stemmed_review = stem_txt(stopwords_removed_review)

    # Transform the preprocessed review using the CountVectorizer
    review_vector = cv.transform([stemmed_review]).toarray()

    # Make the prediction
    prediction = loaded_model.predict(review_vector)

    # Return the sentiment as a string
    if prediction[0] == 1:
        return "Positive"
    else:
        return "Negative"
```

## Create gradio interface


Use the `gradio.Interface` class to create a web interface with a text input field and a text output field, linked to the prediction function.


```python
import gradio as gr

# Create the Gradio interface
iface = gr.Interface(
    fn=predict_sentiment,
    inputs=gr.Textbox(label="Enter movie review"),
    outputs=gr.Textbox(label="Predicted sentiment")
)
```
